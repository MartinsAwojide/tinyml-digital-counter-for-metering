```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
image_transform = cv2.imread("Img_1_473385748.bmp", cv2.IMREAD_GRAYSCALE)
digits = 5

for digit in range(digits):
    if digit >= 0: 
        initial_x = 18
        x = initial_x + digit*15 
        y = 14
        w = 15
        h = 50
        print(f"digit:{digit},x:{x},y:{y},w:{w},h:{h}")
        cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
        plt.imshow(cropped)
        plt.show()
```

```python
image_transform = cv2.imread("Img_1_473385748.bmp", cv2.IMREAD_GRAYSCALE)
digits_before = 4
digit_after = 2

# Before dp 
for digit in range(digits_before):
    if digit >= 0: 
#         if digit == 3 or digit ==4: 
#             continue 
        initial_x = 18
        x = initial_x + digit*15 
        y = 14
        w = 15
        h = 50
        print(digit)
        cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
        # cropped = img[top_edge : bottom_edge, left_edge : right_edge]
        cropped = cropped[y:y+h, x:x+w]
        cv2.imwrite(f'cropped/{digit}_c.png',cropped)
        plt.imshow(cropped)
        plt.show()

# After dp
for digit in range(digit_after):
    if digit >= 0:
        initial_x = 88
        x = initial_x + digit*15 
        y = 14
        w = 15
        h = 50
        print(digit)
        cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
        cropped = cropped[y:y+h, x:x+w]
        cv2.imwrite(f'cropped/{digit}_c.png',cropped)
        plt.imshow(cropped)
        print(cropped.shape)
        plt.show()
```

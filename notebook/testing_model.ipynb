```python
import tensorflow as tf 
import numpy as np 
# import matplotlib.pyplot as plt 
```

```python
# Location of tflite model file (float32 or int8 quantized)
model_path = "model/trained.tflite"
```

```python
# Load TFLite model and allocate tensors.
interpreter = tf.lite.Interpreter(model_path=model_path)
```

```python
# Get input and output tensors.
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

# Allocate tensors
interpreter.allocate_tensors()

# Print the input and output details of the model
print()
print("Input details:")
print(input_details)
print()
print("Output details:")
print(output_details)
print()
```

```python
images = "finished/0_085a4e30-a96a-46ca-b188-af562d78caff.png"
size = 23 
```

```python
image_open = open(images, 'rb')
read_image = image_open.read()
```

```python
read_image
```

```python
image_decode = tf.image.decode_jpeg(read_image)
print("This is the size of the Sample image:",image_decode.shape, "\n")
# print("This is the array for Sample image:",image_decode)
```

```python
resize_image = tf.image.resize(image_decode, [23, 23])
print("This is the Shape of resized image",resize_image.shape)
# print("This is the array for resize image:",resize_image)
```

```python
resize_image.shape
```

```python
resize_image = np.array(resize_image, dtype=np.int8)
```

```python
# If the expected input type is int8 (quantized model), rescale data
input_type = input_details[0]['dtype']
print(input_type)
if input_type == np.int8:
    input_scale, input_zero_point = input_details[0]['quantization']
    print("Input scale:", input_scale)
    print("Input zero point:", input_zero_point)
    print()
    np_features = (resize_image / input_scale) + input_zero_point
    np_features = np.around(resize_image)
    print(np_features.dtype)
```

```python
resize_image.dtype
```

```python
# Add dimension to input sample (TFLite model expects (# samples, data))
np_features = tf.expand_dims(resize_image, axis=0)
```

```python
np_features.shape
```

```python
np_features.dtype
```

```python
np_features = tf.image.grayscale_to_rgb(
    np_features, name=None
)
```

```python
# Create input tensor out of raw features
interpreter.set_tensor(input_details[0]['index'], np_features)

# Run inference
interpreter.invoke()
```

```python
# output_details[0]['index'] = the index which provides the input
output = interpreter.get_tensor(output_details[0]['index'])
```

```python
output
```

```python
# If the output type is int8 (quantized model), rescale data
output_type = output_details[0]['dtype']
if output_type == np.int8:
    output_scale, output_zero_point = output_details[0]['quantization']
    print("Raw output scores:", output)
    print("Output scale:", output_scale)
    print("Output zero point:", output_zero_point)
    print()
    output = output_scale * (output.astype(np.float32) - output_zero_point)

# Print the results of inference
print("Inference output:", output)
```

```python
predicted = tf.argmax(output[0])
predicted
```

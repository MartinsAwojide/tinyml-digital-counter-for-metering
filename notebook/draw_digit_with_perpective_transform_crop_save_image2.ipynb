## Perpective Transformation

```python
import cv2
from operator import itemgetter
from glob import glob
image = cv2.imread('Img_4_473386053.bmp', cv2.IMREAD_GRAYSCALE)
# Coordinates that you want to Perspective Transform
pts1 = np.float32([[87,87],[210,87],[87,127],[210,127]])
# Size of the Transformed Image
pts2 = np.float32([[0,0],[230,0],[0,80],[230,80]])
for val in pts1:
    cv2.circle(image,(val[0],val[1]),5,(0,255,0),-1)
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(image,M,(230,80))
cv2.imwrite("transformed2.jpg", dst)
plt.imshow(dst)
```

## Crop and Save images

```python
import os 
import uuid 
```

```python
path = "data2/"

for file in os.listdir(path): 
    gen = str(uuid.uuid4())
    print(file+gen)
```

```python
# Before
!ls finished | wc -l
```

```python
digits_before = 4
digit_after = 2

for file in os.listdir(path): 
    image_transform = cv2.imread("data2/" +file, cv2.IMREAD_GRAYSCALE)
    print(image_transform)

    # Before dp 
    for digit in range(digits_before):
        if digit >= 0: 
            if digit == 3 or digit ==4: 
                continue 
            initial_x = 47
            x = initial_x + digit*15 
            y = 15
            w = 15
            h = 25
            gen = str(uuid.uuid4())
            print(digit)
            cropped = image_transform
            cropped = cropped[y:y+h, x:x+w]
            cv2.imwrite(f'finished/{digit}_{gen}.png',cropped)
            plt.imshow(cropped)
            plt.show()

    # After dp
    for digit in range(digit_after):
        if digit >= 0:
            initial_x = 89
            x = initial_x + digit*15 
            y = 15
            w = 15
            h = 25
            gen = str(uuid.uuid4())
            print(digit)
            cropped = image_transform 
            cropped = cropped[y:y+h, x:x+w]
            cv2.imwrite(f'finished/{digit}_{gen}.png',cropped)
            plt.imshow(cropped)
            plt.show()
```

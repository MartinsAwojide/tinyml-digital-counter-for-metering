```python
import cv2
from operator import itemgetter
from glob import glob
# paper = cv2.imread('./Photos/book.jpg')
image = cv2.imread('Img_4_473386053.bmp', cv2.IMREAD_GRAYSCALE)
# Coordinates that you want to Perspective Transform
pts1 = np.float32([[87,87],[210,87],[87,127],[210,127]])
# Size of the Transformed Image
# pts2 = np.float32([[0,0],[500,0],[0,400],[500,400]])
pts2 = np.float32([[0,0],[230,0],[0,80],[230,80]])
for val in pts1:
    cv2.circle(image,(val[0],val[1]),5,(0,255,0),-1)
M = cv2.getPerspectiveTransform(pts1,pts2)
# dst = cv2.warpPerspective(image,M,(500,400))
dst = cv2.warpPerspective(image,M,(230,80))
cv2.imwrite("transformed2.jpg", dst)
plt.imshow(dst)
```

```python
import cv2
from operator import itemgetter
from glob import glob
# paper = cv2.imread('./Photos/book.jpg')
image = cv2.imread('Img_24_473387381.bmp', cv2.IMREAD_GRAYSCALE)
# Coordinates that you want to Perspective Transform
pts1 = np.float32([[15,20],[240,15],[15,90],[240,80]])
# Size of the Transformed Image
# pts2 = np.float32([[0,0],[500,0],[0,400],[500,400]])
pts2 = np.float32([[0,0],[230,0],[0,80],[230,80]])
for val in pts1:
    cv2.circle(image,(val[0],val[1]),5,(0,255,0),-1)
M = cv2.getPerspectiveTransform(pts1,pts2)
# dst = cv2.warpPerspective(image,M,(500,400))
dst = cv2.warpPerspective(image,M,(230,80))
cv2.imwrite("transformed.jpg", dst)
plt.imshow(dst)
```

```python
image_transform = cv2.imread("transformed.jpg", cv2.IMREAD_GRAYSCALE)
digits = 8

for digit in range(digits):
    x = digit*20# + 10
    y = 35
    w = 20
    h = 45
    print(digit)
    cropped = cv2.rectangle(image_transform,(x,y),(x+w,y+h),(0,255,0),1)
    plt.imshow(cropped)
    plt.show()
```
